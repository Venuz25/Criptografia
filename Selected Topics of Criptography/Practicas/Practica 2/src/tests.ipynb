```python
import time
import math
```

```python
# Funcion para sumar dos puntos en una EC sobre Zp
def sumaPuntos(P, Q, p):

    if P == (0, 1, 0):
        return Q
    if Q == (0, 1, 0):
        return P

    if Q[0]-P[0] == 0:
        return (0, 1, 0) # Punto en el infinito

    num = (Q[1] - P[1]) % p
    den = (Q[0] - P[0]) % p
    
    try:
        inv_den = pow(den, -1, p)
    except ValueError:
        return (0, 1, 0) # Si el inverso no existe, es punto en el infinito

    pendiente = (num * inv_den) % p

    x3 = (pendiente**2 - P[0] - Q[0]) % p
    y3 = (pendiente * (P[0] - x3) - P[1]) % p
    return (x3, y3, 1)
```

```python
def sum2P(P, a, p):
    if P == (0, 1, 0):
        return (0, 1, 0) # Punto en el infinito

    try:
        inv = pow(2 * P[1], -1, p)
    except ValueError:
        return (0, 1, 0) # Si el inverso no existe, es punto en el infinito
    
    pendiente = (3 * P[0]**2 + a) * inv % p

    x3 = (pendiente**2 - 2 * P[0]) % p
    y3 = (pendiente * (P[0] - x3) - P[1]) % p
    return (x3, y3, 1)
```

```python
# Función kP
def kP(a, b, p, k, P):
    if k == 1: return P
    elif k % 2 == 0: return sum2P(kP(a, b, p, k // 2, P), a, p)
    else: return sumaPuntos(P, kP(a, b, p, k - 1, P), p)
```

```python
def logartmoDiscretoEC(p, a, b, G, P):
    limite = p + int(2 * math.sqrt(p)) + 2
    if P == (0, 1, 0):
        return 0
        
    for k in range(1, limite):
        if kP(a, b, p, k, G) == P:
            return k
    return -1
```
