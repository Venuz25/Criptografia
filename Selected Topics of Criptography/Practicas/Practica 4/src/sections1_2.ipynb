```python
import random
from Crypto.Util.number import getPrime, isPrime
from sympy import randprime
```

```python
def genPrimos():
  q = randprime(11,  1025)
  k = 2
  p = 0
  while isPrime(p) == False:
    p = k*q + 1
    k += 1
  return p, q
```

```python
def generador(p, q):
    e = (p - 1) // q

    while True:
        h = random.randint(1, p - 1)
        g = pow(h, e, p)

        if 2 <= g <= p - 1:
            if pow(g, q, p) == 1:
                return g
```

```python
def keysDSA():
  p, q = genPrimos()
  g = generador(p, q)

  d = random.randint(0, q)
  x = pow(g, d, p)

  return p, q, g, d, x
```

```python
p, q, g, d, x = keysDSA()
print(f"p: {p}")
print(f"q: {q}")
print(f"g: {g}")
print(f"d: {d}")
print(f"x: {x}")
```

```python
def firma(m, p, q, g, d):
  K_E = random.randint(1, q-1)
  r = pow(g, K_E, p) % q
  s = (m + d * r) * pow(K_E, -1, q) % q
  return r, s
```
