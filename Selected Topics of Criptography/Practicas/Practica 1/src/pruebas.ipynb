```python
from Crypto.Util.number import getPrime, isPrime
import random
```

```python
# Funcion que encuentra los valores de a y b para una EC no singular sobre Zp
# y^2 = x^3 + ax + b
def curvasNoSingulares(p, f, t):
    noSing = []
    for a in range(p):
        for b in range(p):
            if (4 * a**3 + 27 * b**2) % p != 0:
                noSing.append((a, b))
                if f == 1 and len(noSing) == t:
                    return noSing
    return noSing
```

```python
p = getPrime(2048)
t = 10000
print("p:", p)
random.choice(curvasNoSingulares(p, 1, t))
```

```python
def curvasSingulares(p):
    sing = []
    for a in range(p):
        for b in range(p):
            if (4 * a**3 + 27 * b**2) % p == 0:
                sing.append((a, b))
    return sing
```

```python
# Funcion para encontrar los puntos racionales en una EC
def encontrarPuntosEC(a, b, p):
    puntos = []
    for x in range(p):
        y2 = (x**3 + a * x + b) % p
        for y in range(p):
            if (y * y) % p == y2:
                puntos.append((x, y, 1))
    puntos.append((0, 1, 0)) # Punto en el infinito
    return puntos
```

```python
# Funcion para sumar dos puntos en una EC sobre Zp
def sumaPuntos(P, Q, p):

    if Q[0]-P[0] == 0:
        return (0, 1, 0) # Punto en el infinito

    num = (Q[1] - P[1]) % p
    den = (Q[0] - P[0]) % p
    
    try:
        inv_den = pow(den, -1, p)
    except ValueError:
        return (0, 1, 0) # Si el inverso no existe, es punto en el infinito

    pendiente = (num * inv_den) % p

    x3 = (pendiente**2 - P[0] - Q[0]) % p
    y3 = (pendiente * (P[0] - x3) - P[1]) % p
    return (x3, y3, 1)
```

```python
def sum2P(P, a, p):
    if P == (0, 1, 0):
        return (0, 1, 0) # Punto en el infinito

    try:
        inv = pow(2 * P[1], -1, p)
        print("Inverso:", inv)
    except ValueError:
        return (0, 1, 0) # Si el inverso no existe, es punto en el infinito
    
    pendiente = (3 * P[0]**2 + a) * inv % p
    print("Pendiente:", pendiente)

    x3 = (pendiente**2 - 2 * P[0]) % p
    y3 = (pendiente * (P[0] - x3) - P[1]) % p
    return (x3, y3, 1)
```

```python
def noPuntosPrimosEC(p):
    noPuntos = []
    curvas = curvasNoSingulares(p, 0)
    for curva in curvas:
        a, b = curva
        puntos = encontrarPuntosEC(a, b, p)
        t = len(puntos)
        if isPrime(t):
            noPuntos.append([(a, b), t])
    return noPuntos
```
