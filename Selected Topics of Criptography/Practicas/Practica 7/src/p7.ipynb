```python
from Crypto.PublicKey import RSA
from Crypto.Util.number import getPrime, inverse
```

```python
p = getPrime(512)
q = getPrime(512)

print("p: ", p)
print("q: ",q)
```

```python
n = p * q
pi = (p - 1) * (q - 1) 
print("n: ",n)
print("pi(n):  ", pi)
```

```python
e = inverse(pi, n)
print("e: ", e)
```

```python
d = inverse(e, pi)
print("d: ", d)
```
