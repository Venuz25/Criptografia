```python
def euclid(a,b):
    if a % b == 0:
        return b
    else:
        return euclid(b, a % b)
```

```python
a = 5
b = 17
print(euclid(a,b))
```

```python
def extendedX_rec(a,b):
    if a %  b == 0:
        return (0, 1 , b)
    else:
        A, B, y = extendedX_rec(b, a % b)
        return (B, A - B * (a // b), y)
```

```python
a = 5
b = 17

m, n, y = extendedX_rec(a,b)
print(f"{m}*{a} + {n}*{b} = {y}")
print(m*a + n*b)
```

```python
def fi(n):
    count = 0
    for k in range(1, n):
        if euclid(k, n) == 1:
            count += 1
    return count
```

```python
n = 12
print(fi(n))
```
