```python
def inserta(x, lst, i):
    return lst[:i] + [x] + lst[i:]
```

```python
def inserta_multiple(x, lst):
    return [inserta(x, lst, i) for i in range(len(lst)+1)]
```

```python
def permuta(c):
    if len(c) == 0:
        return [[]]
    lst = []
    for i in permuta(c[1:]):
        lst = lst + inserta_multiple(c[0], i)
    return lst
```

```python
c = [1, 2, 3]
print(permuta(c))
```

```python
def sgn(p):
    count = 0
    i = -1
    a = []

    for k in range (len(p)):
        a = a + [p[k]]
    while i < len(a)-2:
        i=i+1
        if a[i] > a[i+1]:
            a[i], a[i+1] = a[i+1], a[i]
            count = count + 1
            i = -1
            continue
    if count %2 == 0:
        return 1
    else:
        return -1
```

```python
def det(a):
    n = len(a)
    s = range(n)
    t = permuta(s)
    d = 0
    for u in t:
        r = sgn(u)
        for i in range(n):
            r = r * a[i , u[i]]
        d = d + r
    return d
```

```python
import numpy as np
a = [[1,3],[2,4]]
a = np.array(a)
d = np.linalg.det(a)
print(d)
```

```python
def detK(a, k):
    n = len(a)
    b = np.zeros(((n-1), (n-1)))
    for i in range(1,n):
        for j in range(n):
            if j < k:
                b[i-1, j] = a[i, j]
            elif j > k:
                b[i-1, j-1] = a[i, j]
    return b
```
